```python
import numpy as np
import matplotlib.pyplot as plt
```

## Zad.1 

x' = rx(t)ln $( \frac{K}{x(t)})$

x(75) = 10 

```python
t = np.arange(75.0, 125.1, 0.001)
K = 100000
h = 0.001
r = 0.4
N1 = [10]

#Gompertz 
for _ in range(len(t)):
    new1 = N1[-1] + h * r * N1[-1] * np.log(K / N1[-1])

    N1.append(new1)


#model Verhulsta
lamb = 0.4
N2 = [10]
ver_bis = lamb**2 * N1[-1] * (1 - (N1[-1]/K)) * ( 1 - 2*N1[-1]/K)

for _ in range(len(t)):
    ver = N2[-1] + h * lamb * (1 - N2[-1]/K) * N2[-1] + ((h**2) / 2) * ver_bis
    N2.append(ver)

#wykres 
plt.plot(t, N1[:-1], label = 'Gompertz')
plt.xlabel("t")
plt.ylabel("rozmiar guza")
plt.plot(t, N2[:-1], label = 'Verhulst')
plt.legend()
plt.show()
```

## Zad.2 

### podpunkt a 

```python
#podpunkt a 
t = np.arange(0, 10, 0.0001)
h = 0.0001
N1 = [3]
N2 = [4]

e1 = 1.25
y1 = 0.5
h1 = 0.1

e2 = 0.5
y2 = 0.2
h2 = 0.2

for _ in range(len(t)):
    new1 = N1[-1] + h * (e1 - y1 * (h1*N1[-1] + h2*N2[-1])) * N1[-1]
    new2 = N2[-1] + h * (e2 - y2 * (h1*N1[-1] + h2*N2[-1])) * N2[-1]

    N1.append(new1)
    N2.append(new2)

#wykres
plt.plot(t, N1[:-1], label = 'Gatunek 1')
plt.plot(t, N2[:-1], label = 'Gatunek 2')
plt.xlabel("t")
plt.ylabel("liczebnoc gatunku")
plt.legend()
plt.show()
```

Oba gatunki osigaj stan rwnowagi, jednak stan ten jest wyszy u gatunku 1; dominuje nad gatunkiem 2. 

### podpunkt b

```python

t = np.arange(0, 10, 0.0001)
h = 0.0001

N1 = [3]
N2 = [4]

e1 = 5
y1 = 4
h1 = 1

e2 = 5
y2 = 8
h2 = 4

for _ in range(len(t)):
    new1 = N1[-1] + h * (e1 - y1 * (h1*N1[-1] + h2*N2[-1])) * N1[-1]
    new2 = N2[-1] + h * (e2 - y2 * (h1*N1[-1] + h2*N2[-1])) * N2[-1]

    N1.append(new1)
    N2.append(new2)


#wykres
plt.plot(t, N1[:-1], label = 'Gatunek 1')
plt.plot(t, N2[:-1], label = 'Gatunek 2')
plt.xlabel("t")
plt.ylabel("liczebnoc gatunku")
plt.legend()
plt.show()
```

Gatunek 1 z pocztku ma mniejsz liczebno od gatunku 2, jednak z czasem dominuje nad gatunkiem 2; osiga wyszy stan rwnowagi. 

W obu przypadkach gatunek 1 dominuje nad gatunkiem 2. W podpunkcie a populacja ronie po czym si stabilizuje, natomiast w podpunkcie b, populacja maleje i si stabilizuje.

### podpunkt c, d, e

```python
t = np.arange(0, 10, 0.0001)
h = 0.0001

e1 = 0.8
y1 = 1
h1 = 0.3

e2 = 0.4
y2 = 0.5
h2 = 0.4


#c
N1 = [4]
N2 = [8]

for _ in range(len(t)):
    new1 = N1[-1] + h * (e1 - y1 * (h1*N1[-1] + h2*N2[-1])) * N1[-1]
    new2 = N2[-1] + h * (e2 - y2 * (h1*N1[-1] + h2*N2[-1])) * N2[-1]

    N1.append(new1)
    N2.append(new2)


#d)
N3 = [8]
N4 = [8]

for _ in range(len(t)):
    new3 = N3[-1] + h * (e1 - y1 * (h1*N3[-1] + h2*N4[-1])) * N3[-1]
    new4 = N4[-1] + h * (e2 - y2 * (h1*N3[-1] + h2*N4[-1])) * N4[-1]
    
    N3.append(new3)
    N4.append(new4)


#e)
N5 = [12]
N6 = [8]


for _ in range(len(t)):
    new5 = N5[-1] + h * (e1 - y1 * (h1*N5[-1] + h2*N6[-1])) * N5[-1]
    new6 = N6[-1] + h * (e2 - y2 * (h1*N5[-1] + h2*N6[-1])) * N6[-1]

    N5.append(new5)
    N6.append(new6)


grid = np.meshgrid(np.arange(0, 13, 0.4), np.arange(0, 13, 0.4))

N1_prim = (e1 - y1 * (h1 * grid[0] + h2 * grid[1])) * grid[0]
N2_prim = (e2 - y2 * (h1 * grid[0] + h2 * grid[1])) * grid[1]

N3_prim = (e1 - y1 * (h1 * grid[0] + h2 * grid[1])) * grid[0]
N4_prim = (e2 - y2 * (h1 * grid[0] + h2 * grid[1])) * grid[1]

N5_prim = (e1 - y1 * (h1 * grid[0] + h2 * grid[1])) * grid[0]
N6_prim = (e2 - y2 * (h1 * grid[0] + h2 * grid[1])) * grid[1]


#wykresy 
plt.plot(N1, N2, label = '(N1, N2) = (4,8)')
plt.plot(N3, N4, label = '(N1, N2) = (8,8)')
plt.plot(N5, N6, label = '(N1, N2) = (12,8)')

plt.quiver(grid[0], grid[1], N1_prim, N2_prim)
plt.quiver(grid[0], grid[1], N3_prim, N4_prim)
plt.quiver(grid[0], grid[1], N5_prim, N6_prim)

plt.xlabel("gatunek 1")
plt.ylabel("gatunek 2")

plt.legend()
plt.show()
```

Im wiksza pocztkowa liczebno pierwszego gatunku, tym pocztkowy spadek liczebnoci tego gatunku jest szybszy.
